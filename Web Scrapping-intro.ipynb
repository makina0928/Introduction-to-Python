## Beautiful Soup

```python
import os
os.getcwd()
```

```python
ls
```

```python
filename ='html-2'
```

```python
f= open(filename,'r')
contents=f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(contents, 'html.parser')
type(soup)
```

```python
soup.title.text
```

```python
print(soup.prettify())
```

```python
soup.head.text
```

```python
soup.div.text
```

```python
spanElement = soup.find('div')
if spanElement != None:
    print(spanElement.text)
```

```python
soup.find('section',{'id':"content"})
```

```python
divElements = soup.find_all('div')
divElements
```

```python
for element in divElements:
    text = element.text
    if text != None:
        print(text)
```

```python
divElements[0].text
```

```python
ContentElement = soup.find('div',{'class':"container"})
Contentdiv = ContentElement.find('section')
siblingDiv = Contentdiv.find_next_sibling('section')
print(Contentdiv.text,siblingDiv.text)
```
